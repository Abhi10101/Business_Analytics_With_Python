```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum() #giving sum of isnull
```

```python
df.dropna(inplace=True) #dropping null val
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df.describe()
```

# Exploratory Data Analysis


gender 

```python
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)

```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

### we can see females purchased more and their spending power is also more than males

# Age

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)

```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

#### most of buyers are from age group 26-35 and are females


# state

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
# totalamount /sales from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.barplot(x='State',y='Amount',data=sales_state)
```

### most orders are from UP and Maharastra 

## Marital Status

```python
ax=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(8,1)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

```python
# we see that most of buyers are married women and have high purchasing power
```

Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Occupation',data=df,)

for bars in ax.containers:
    ax.bar_label(bars)

```

```python
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Occupation',y='Amount')
```

most buyers are from IT industry

product category

```python
sns.set(rc={'figure.figsize':(30,5)})
ax=sns.countplot(x='Product_Category',data=df,)

for bars in ax.containers:
    ax.bar_label(bars)

```

```python
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(35,5)})
sns.barplot(data=sales_state,x='Product_Category',y='Amount')
```

most sold products are food and footwear

```python
sales_state=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(12,7)})
sns.barplot(data=sales_state,x='Product_ID',y='Orders')
```
